## 35. Search Insert Position. Easy

Given a sorted array of distinct integers and a target value, return the index if the target is found. If not, return the index where it would be if it were inserted in order.

You must write an algorithm with O(log n) runtime complexity.

 

Example 1:

Input: nums = [1,3,5,6], target = 5
Output: 2

Example 2:

Input: nums = [1,3,5,6], target = 2
Output: 1

Example 3:

Input: nums = [1,3,5,6], target = 7
Output: 4

Constraints:

1 <= nums.length <= 104
-104 <= nums[i] <= 104


nums contains distinct values sorted in ascending order.

-104 <= target <= 104

```python
# increase speed
class Solution:
    def searchInsert(self, nums: List[int], target: int) -> int:
        start_position = 0
        end_position = len(nums) - 1
#         if target > nums[end_position]:
#             return end_position + 1
        for i, _ in enumerate(nums):
            mid_ind = start_position + (end_position - start_position) // 2
            mid_elem = nums[mid_ind]
            if target == mid_elem:
                return mid_ind
            elif target > mid_elem:
                start_position = mid_ind + 1
            else:
                end_position = mid_ind - 1 
        return start_position
```

```python
%%time
sol = Solution()
nums = [1,3,5,6]
target = 7
sol.searchInsert(nums, target)
```
