# 283. Move Zeroes. Easy

Given an integer array nums, move all 0's to the end of it while maintaining the relative order of the non-zero elements.

Note that you must do this in-place without making a copy of the array.

 

Example 1:

Input: nums = [0,1,0,3,12]
Output: [1,3,12,0,0]

Example 2:

Input: nums = [0]
Output: [0]

```python
# from typing import List
# class Solution:
#     def moveZeroes(self, nums: List[int]) -> None:
#         """
#         Do not return anything, modify nums in-place instead.
#         """
#         first = 0
#         last = len(nums) - 1
#         while first <= last:
#             if nums[first] == 0:
#                 nums[first], nums[last] = nums[last], nums[first]
#                 last -= 1
#             else:
#                 first += 1
#         print(nums)
from typing import List
class Solution:
    def moveZeroes(self, nums: List[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        first = 0
        for last in range(1, len(nums)):
            if nums[first] == 0 and nums[last] != 0:
                nums[first], nums[last] = nums[last], nums[first]
                
            if nums[first] != 0:
                first += 1
```

```python
%%time
nums = [0,1,0,3,12]
sol = Solution()
sol.moveZeroes(nums) #Output: [1,3,12,0,0]
```

```python
%%time
nums = [0]
sol = Solution()
sol.moveZeroes(nums) #Output: [0]
```

```python
%%time
nums = [0, 1, 0, 3, 2]
sol = Solution()
sol.moveZeroes(nums) #Output: [1, 2, 3, 0, 0]
```
